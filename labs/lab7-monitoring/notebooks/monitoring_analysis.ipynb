# Lab 7 — Inspect captures & monitoring reports

Run cells after invoking `simulate_drift.py`. Export `BUCKET_NAME` before running cells.

> Adjust `REGION`/`PREFIX_CAPTURE`/`PREFIX_REPORTS` when using custom prefixes.

```python
import json
import os

import boto3

REGION = "us-east-1"
bucket = os.environ["BUCKET_NAME"]  # export beforehand
PREFIX_CAPTURE = "lab7/data-capture/"
PREFIX_REPORTS = "lab7/reports/"

s3 = boto3.client("s3", region_name=REGION)
```

```python
def peek_keys(prefix, limit=15):
    resp = s3.list_objects_v2(Bucket=bucket, Prefix=prefix, MaxKeys=limit)
    return [o["Key"] for o in resp.get("Contents", [])]

capture_keys = peek_keys(PREFIX_CAPTURE)
report_keys = peek_keys(PREFIX_REPORTS)
print("Captures:\n", json.dumps(capture_keys, indent=2))
print("Reports:\n", json.dumps(report_keys, indent=2))
```
