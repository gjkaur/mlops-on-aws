# Lab 5 — Pipeline analysis (optional)

Inspect recent **SageMaker Pipeline** executions with boto3. Set `REGION` / `PIPELINE_NAME` below, or infer from **`scripts/pipeline_name.txt`**.

```python
from pathlib import Path
import boto3

REGION = "us-east-1"
script_dir = Path.cwd().resolve().parent / "scripts"
name_file = script_dir / "pipeline_name.txt"
PIPELINE_NAME = (
    name_file.read_text().strip() if name_file.is_file() else "YOUR_PIPELINE_NAME"
)

sm = boto3.client("sagemaker", region_name=REGION)
resp = sm.list_pipeline_executions(PipelineName=PIPELINE_NAME, MaxResults=5)
for ex in resp.get("PipelineExecutionSummaries", []):
    print(ex["PipelineExecutionArn"], ex["PipelineExecutionStatus"])
```
