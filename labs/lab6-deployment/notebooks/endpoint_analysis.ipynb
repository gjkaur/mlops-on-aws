# Lab 6 — Endpoint analysis (optional)

Inspect an endpoint with boto3 (`describe_endpoint`, CloudWatch invocation metrics).

```python
from pathlib import Path
import boto3

REGION = "us-east-1"
p = Path("../scripts/endpoint_name.txt")
endpoint = p.read_text().strip() if p.is_file() else "YOUR_ENDPOINT_NAME"
sm = boto3.client("sagemaker", region_name=REGION)
print(sm.describe_endpoint(EndpointName=endpoint))
```
