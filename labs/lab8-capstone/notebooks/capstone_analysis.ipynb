- Export **`ARTIFACTS_BUCKET_NAME`** before lookups.
- Inspect training artefacts under **`pipeline-training-output/`** prefixes after executions.


```python
import os

import boto3

REGION = "us-east-1"
artifacts = os.environ["ARTIFACTS_BUCKET_NAME"]

s3 = boto3.client("s3", region_name=REGION)
resp = s3.list_objects_v2(Bucket=artifacts, Prefix="pipeline-training-output/", MaxKeys=25)
[o["Key"] for o in resp.get("Contents", [])]
```
